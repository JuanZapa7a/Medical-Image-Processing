## 1. Introduction

A database for deep learning typically refers to a *specialized repository* of data structured and optimized for training and evaluating deep learning models. These databases differ from traditional databases in several key ways:

1. **Large-scale and Diverse Data**: Deep learning databases often contain *massive amounts of data*, which is crucial for training models with high accuracy. This data is *usually diverse*, encompassing various types such as images, text, audio, or video.

2. **Annotated and Labeled Data**: Data in deep learning databases is meticulously *annotated and labeled*. Annotations provide *ground truth* information essential for supervised learning tasks, enabling models to learn from labeled examples.

3. **High Dimensionality and Complexity**: Deep learning datasets can have *high dimensionality*, especially in the case of images and videos where each *data point (e.g., pixel or frame) contributes* to the overall *complexity* of the dataset.

4. **Optimized for Performance**: These databases are optimized for performance to handle large-scale data processing efficiently. They often leverage *distributed computing frameworks* and *storage systems* to manage and process data in parallel.

5. **Version Control and Management**: Due to the dynamic nature of data in deep learning research, *version control and management of datasets are critical*. Tools and practices are employed *to track changes, ensure data integrity, and facilitate reproducibility in experiments*.

6. **Privacy and Security Considerations**: Deep learning datasets may contain *sensitive information*, necessitating robust privacy and security measures to protect data confidentiality and prevent unauthorized access.

7. **Community and Accessibility**: Many deep learning databases are *open-source or publicly accessible*, fostering collaboration and enabling researchers worldwide to contribute, use, and build upon existing datasets.

In essence, a database for deep learning serves as a foundational resource for training, validating, and benchmarking deep learning models. Its specialized characteristics cater to the unique requirements of machine learning algorithms, facilitating advancements in artificial intelligence research and applications.

Leukaemia, which is a type of *blood cancer*, is a group of diseases that affect the blood cells and bone marrow. There are several types of leukaemia, the two main ones being *acute myeloid leukaemia* (AML) and *acute lymphoblastic leukaemia* (ALL). In addition to these, there are other subtypes of leukaemia, including *chronic myeloid leukaemia* (CML) and *chronic lymphocytic leukaemia* (CLL).

Main Types of Leukemia:

1. **Acute Myeloid Leukemia (AML):**
   - It is a type of leukemia that affects immature myeloid cells.
   - It can progress rapidly if left untreated.

2. **Acute Lymphoblastic Leukemia (ALL):**
   - It affects immature lymphoid cells.
   - It is more common in children but can also affect adults.

3. **Chronic Myeloid Leukemia (CML):**
   - It is characterized by uncontrolled growth of myeloid cells.
   - It tends to progress more slowly.

4. **Chronic Lymphocytic Leukemia (CLL):**
   - It involves excessive growth of mature lymphocytic cells.
   - It is more common in adults and generally progresses slowly.

### Causes of Leukemia:

The exact causes of leukemia are not always clear, but some risk factors and potential causes include:

1. **Genetics:**
   - Some forms of leukemia have a genetic component, meaning there may be a hereditary predisposition.

2. **Exposure to Radiation:**
   - Exposure to high levels of ionizing radiation can increase the risk of leukemia.

3. **Exposure to Chemicals:**
   - Prolonged exposure to certain chemicals, such as benzene, can increase the risk.

4. **Genetic Disorders:**
   - Some genetic disorders, such as Down syndrome, can increase the risk of leukemia.

5. **Medical Treatments:**
   - Radiation therapy and certain chemotherapy treatments for other types of cancer can increase the risk of developing leukemia.

It is important to note that while these factors may increase the risk, they do not guarantee that a person will develop leukemia. The disease is complex and can affect individuals without known risk factors.

### Displaying Images from Dataset

Next code selects a certain number of images from a directory containing images of `Blood_Cancer`, calculates the appropriate layout for subplots based on the number of images, and displays the images in the subplots using matplotlib.

---

**Display some images from dataset**

---

````python
import os
import matplotlib.pyplot as plt
from matplotlib.image import imread

# Directory containing images of Blood_Cancer
directory = "Blood_Cancer"

# List of file names in the directory
files = os.listdir(directory)

# Filter only files with common image extensions (you can adjust as needed)
image_files = [file for file in files if file.lower().endswith(('.png', '.jpg', '.jpeg', '.tiff'))]

# Total number of images to visualize (you can change this value)
num_images = 32

# Calculate the number of rows and columns
# Calculate divisors of num_images
divisors = [i for i in range(1, num_images + 1) if num_images % i == 0]
# Select the two central divisors
center = len(divisors) // 2 - 1 if len(divisors) % 2 == 0 and len(divisors) > 1 else len(divisors) // 2
# Determine the number of rows and columns based on selected divisors
rows = divisors[center]
columns = num_images // rows if num_images > 1 else 1
# Print divisors, number of rows, and number of columns
print(divisors, rows, columns)

# Set up subplots
fig, axs = plt.subplots(rows, columns, figsize=(12, 6))

# Iterate over the images and display them in subplots
for i, image in enumerate(image_files[:num_images]):
  image_path = os.path.join(directory, image)
  img = imread(image_path)

  # Determine the position in the subplot
  row = i // columns
  column = i % columns

  # If there is only one row or one column,
  # then axs will be a single axis, not a matrix
  if rows == 1 or columns == 1:
    axs[i].imshow(img)
    axs[i].axis('off')
    axs[i].set_title(image, fontsize=8)
  else:
    # Display the image in the corresponding subplot
    axs[row, column].imshow(img)
    axs[row, column].axis('off')
    axs[row, column].set_title(image, fontsize=8)

# Adjust the layout and show the visualization
plt.tight_layout()
plt.show()
````

## 7. Determining Physical Characteristics of Cells with Cancer



---
**Uses OpenCV to determine area, perimeter (contour), and moments (centroid)**

---


This code performs the following steps:

1. **Choosing a random image:**
   - Randomly selects an image from the specified directory.

2. **Reading and thresholding the image:**
   - Reads the image in grayscale.
   - Applies inverse binary thresholding.

3. **Finding and sorting contours:**
   - Finds contours in the thresholded image.
   - Prints the number of objects found.
   - Sorts the contours by area from largest to smallest.

4. **Calculating features:**
   - Calculates the area, perimeter, centroid, and bounding box of the largest contour.

5. **Visualization with subplots:**
   - Sets up subplots to display the original image, the thresholded image, the contour with centroid and bounding box, and finally, the image with contours and bounding box drawn using `cv2`.

6. **Printing features:**
   - Prints the area, perimeter, and centroid of the largest contour.


````python
import cv2
import numpy as np
import random
from matplotlib.patches import Rectangle

# Choose a random image
random_image = random.choice(image_files)
random_image_path = os.path.join(directory, random_image)

# Read the image in grayscale
image = cv2.imread(random_image_path, cv2.IMREAD_GRAYSCALE)

# Apply thresholding
_, thresholded = cv2.threshold(image, 128, 255, cv2.THRESH_BINARY_INV)

# Find contours
contours, hierarchy = cv2.findContours(thresholded, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

# Print the number of objects found
num_objects = len(contours)
print(f'Number of objects found: {num_objects}')

# Sort contours by area from largest to smallest
contours = sorted(contours, key=cv2.contourArea, reverse=True)

# Calculate features
area = cv2.contourArea(contours[0])
perimeter = cv2.arcLength(contours[0], True)
moments = cv2.moments(contours[0])
centroid_x = int(moments["m10"] / moments["m00"])
centroid_y = int(moments["m01"] / moments["m00"])
x, y, w, h = cv2.boundingRect(contours[0])

# Create subplots
fig, axs = plt.subplots(1, 4, figsize=(12, 4))

# Show images in subplots
axs[0].imshow(image, cmap='gray')
axs[0].set_title('Original Image')

axs[1].imshow(thresholded, cmap='gray')
axs[1].set_title('Thresholding')

# Draw contour and bounding box on the original image
axs[2].imshow(cv2.cvtColor(image, cv2.COLOR_GRAY2BGR))
axs[2].plot(
    np.append(contours[0][:,0,0], contours[0][0,0,0]),
    np.append(contours[0][:,0,1], contours[0][0,0,1]),
    color='lime', linewidth=2)
rect = Rectangle((x, y), w, h, linewidth=2, edgecolor='red', facecolor='none')
axs[2].add_patch(rect)
axs[2].scatter(centroid_x, centroid_y, color='red', s=100, marker='x')
axs[2].set_title('Contour, Centroid, and BBox')

# Draw contours and bounding box using cv2
cv2.drawContours(image, [contours[0]], -1, (0, 255, 0), 1)
cv2.rectangle(image, (x, y), (x + w, y + h), (0, 0, 255), 1)
axs[3].imshow(image, cmap='gray')
axs[3].set_title('Contours and BBox (cv2)')

# Adjust layout and show visualization
plt.tight_layout()
plt.show()

# Print features
print(f"Area: {area}")
print(f"Perimeter: {perimeter}")
print(f"Centroid: ({centroid_x}, {centroid_y})")
````

>**NOTE**: If `np.append` is not used, the contour will not close upon itself.

Next code prints the x and y coordinates of the points forming the contour in the image. `contours[0][:,0,0]` and `contours[0][:,0,1]` return all the x and y coordinates respectively, while `contours[0][0,0,0]` and `contours[0][0,0,1]` return the x and y coordinates of the first point of the contour.

Apart from **`area, perimeter, and centroid`**, there are several other features that can be extracted or shown from the contour of an object in an image. Some of these features include:

1. **`Compactness`**: Measure of how compact the shape is, calculated as the square of perimeter divided by area. Higher values indicate more irregular shapes.

2. **`Elongation`**: Measure of how elongated or stretched the shape is, calculated as the ratio of the major axis length to the minor axis length of the fitted ellipse.

3. **`Extent`**: Ratio of the contour area to the bounding box area. It measures how much of the bounding box is filled by the object.

4. **`Solidity`**: Ratio of contour area to convex hull area. It measures the proportion of the convex hull that is filled by the object.

5. **`Eccentricity`**: Measure of how much the shape deviates from a perfect circle. It is calculated as the square root of the ratio of the major and minor axes of the fitted ellipse.

6. **`Orientation`**: Angle of the major axis of the fitted ellipse with respect to the x-axis.

7. **`Convexity Defects`**: Points where the contour deviates significantly from the convex hull. These points can indicate concavities or holes in the object.

8. **`Hu Moments`**: A set of seven moments invariant to translation, rotation, and scale. They capture the shape of the object more robustly than raw moments.

9. **`Haralick Texture Features`**: Statistical measures derived from the gray-level co-occurrence matrix (GLCM), describing texture characteristics such as contrast, energy, and entropy.

These features can provide valuable information about the shape, texture, and overall characteristics of objects in an image, which can be useful for various applications such as **`object recognition, classification, and segmentation`**.

````python
import cv2
import numpy as np
import random
from matplotlib.patches import Rectangle

# Choose a random image
random_image = random.choice(image_files)
random_image_path = os.path.join(directory, random_image)

# Read the image in grayscale
image = cv2.imread(random_image_path, cv2.IMREAD_GRAYSCALE)

# Apply thresholding
_, thresholded = cv2.threshold(image, 128, 255, cv2.THRESH_BINARY_INV)

# Find contours
contours, hierarchy = cv2.findContours(thresholded, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
# Print the number of objects found
num_objects = len(contours)
print(f'Number of objects found: {num_objects}')

# Sort contours by area from largest to smallest
contours = sorted(contours, key=cv2.contourArea, reverse=True)

# Calculate features
area = cv2.contourArea(contours[0])
perimeter = cv2.arcLength(contours[0], True)
moments = cv2.moments(contours[0])
centroid_x = int(moments["m10"] / moments["m00"])
centroid_y = int(moments["m01"] / moments["m00"])
x, y, w, h = cv2.boundingRect(contours[0])

# Fit ellipse to the contour
ellipse = cv2.fitEllipse(contours[0])
major_axis_length = max(ellipse[1])
minor_axis_length = min(ellipse[1])
elongation = major_axis_length / minor_axis_length
eccentricity = np.sqrt(1 - (minor_axis_length ** 2) / (major_axis_length ** 2))

# Create subplots
fig, axs = plt.subplots(1, 4, figsize=(12, 4))

# Show images in subplots
axs[0].imshow(image, cmap='gray')
axs[0].set_title('Original Image')

axs[1].imshow(thresholded, cmap='gray')
axs[1].set_title('Thresholding')

# Draw contour and bounding box on the original image
axs[2].imshow(cv2.cvtColor(image, cv2.COLOR_GRAY2BGR))
axs[2].plot(
    np.append(contours[0][:,0,0], contours[0][0,0,0]),
    np.append(contours[0][:,0,1], contours[0][0,0,1]),
    color='lime', linewidth=2)
rect = Rectangle((x, y), w, h, linewidth=2, edgecolor='red', facecolor='none')
axs[2].add_patch(rect)
axs[2].scatter(centroid_x, centroid_y, color='red', s=100, marker='x')
axs[2].set_title('Contour, Centroid, and BBox')

# Draw contours and bounding box using cv2
cv2.drawContours(image, [contours[0]], -1, (0, 255, 0), 1)
cv2.rectangle(image, (x, y), (x + w, y + h), (0, 0, 255), 1)
axs[3].imshow(image, cmap='gray')
axs[3].set_title('Contours and BBox (cv2)')

# Adjust layout and show visualization
plt.tight_layout()
plt.show()

# Print features
print(f"Area: {area}")
print(f"Perimeter: {perimeter}")
print(f"Centroid: ({centroid_x}, {centroid_y})")
print(f"Elongation: {elongation}")
print(f"Eccentricity: {eccentricity}")
````
